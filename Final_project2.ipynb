Sửa giá tị null

```python
fact["Discount_pct"].fillna(0, inplace=True)
fact["Discount_pct"]
```

```python
#Thêm cột revenue: phụ thuộc vào có Discount hay không
def calculate_revenue(i):
  if i["Coupon_Status"] == "Used":
    return (i["Avg_Price"] * i["Quantity"]) * (1 - i["Discount_pct"]) * (1 + i["GST"]) + i["Delivery_Charges"]
  else:
    return (i["Avg_Price"] * i["Quantity"]) * (1 + i["GST"]) + i["Delivery_Charges"]
fact["Revenue"] = fact.apply(calculate_revenue, axis=1)
fact
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Assuming 'fact' is your existing DataFrame
# Assuming 'customer' is your Customer DataFrame with 'CustomerID' and 'Location' columns

# Merge 'fact' with 'customer' on 'CustomerID'
merged2 = fact.merge(df_customer[['CustomerID', 'Location']], how='left', on='CustomerID')

# Reshape data for heatmap
merged2_agg = merged2.groupby(['Month', 'Location']).agg({'Revenue': 'sum'}).reset_index()
heatmap_data = merged2_agg.pivot(index='Month', columns='Location', values='Revenue')

def animate(i):
    # Clear existing plot
    plt.cla()

    # Select data for the current month
    current_month_data = heatmap_data.iloc[i, :]

    # Create heatmap
    plt.pcolor(current_month_data.values, vmin=0, vmax=heatmap_data.max().max())

    # Add labels and title
    plt.xlabel('Location')
    plt.ylabel('Month')
    plt.title(f'Sales by Location - {heatmap_data.index[i]}')

    # Set colorbar
    plt.colorbar()

fig, ax = plt.subplots(figsize=(10, 6))
anim = FuncAnimation(fig, animate, frames=len(heatmap_data.index), interval=200)

# Save animation as a GIF file
anim.save('sales_heatmap_over_time.gif', writer='pillow', fps=10)
```
